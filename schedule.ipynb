```python
import conversions
import manipulator
from pandas import read_excel, DataFrame
```

```python
data = read_excel("data/View_My_Courses.xlsx",skiprows=2)
data
```

```python
data_filtered = data[["Section", "Meeting Patterns"]]
data_filtered
```

```python
data_cleaned = data_filtered.dropna()
data_cleaned
```

```python
data_dict = data_cleaned.to_dict(orient='records')
data_dict
```

```python
data_dict_split = manipulator.split_meeting_patterns(data_dict)
data_dict_split
```

```python
data_as_dataframe = DataFrame(data_dict_split)

data_as_dataframe
```

```python
data_as_dataframe["Section"] = data_as_dataframe["Section"].apply(conversions.get_name)

data_as_dataframe
```

```python
data_as_dataframe["Meeting Patterns"] = data_as_dataframe["Meeting Patterns"].astype(str).apply(conversions.handle_meeting_patterns)
data_as_dataframe
```
